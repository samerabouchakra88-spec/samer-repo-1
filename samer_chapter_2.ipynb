```python
import seaborn as sns
import matplotlib.pyplot as plt

# Drop rows where 'Data_Value' is NaN for cleaner visualization
violin_df = heart_df.dropna(subset=['Data_Value']).copy()

# Create the violin plot
plt.figure(figsize=(12, 7))
sns.violinplot(x='Stratification2', y='Data_Value', data=violin_df, palette='viridis')
plt.title('Distribution of Heart Disease Mortality (Data_Value) by Stratification2')
plt.xlabel('Stratification2')
plt.ylabel('Data Value (per 100,000 population)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
