```python
توقيف_df = pd.read_excel('/جدول باسماء الموقوفين الصادر بحقهم مذكرات.xls', header=5)

# Get the new column names from the first row and strip whitespace
new_columns = [str(col).strip() for col in توقيف_df.iloc[0]]
توقيف_df.columns = new_columns

# Drop the first row which is now the header
توقيف_df = توقيف_df[1:]

# Now set the index
توقيف_df = توقيف_df.set_index('م')
توقيف_df.head()
```

```python
توقيف_df.describe()
```

```python
# Find duplicate rows based on all columns
duplicate_rows = توقيف_df[توقيف_df.duplicated()]

print("Duplicate rows based on all columns:")
if not duplicate_rows.empty:
    display(duplicate_rows)
else:
    print("No duplicate rows found based on all columns.")
```

It's also useful to check for duplicates in specific columns, such as 'الاسم الثلاثي' (Full Name), as some individuals might have the same name but different details, or true duplicates might exist.

```python
print(f"Shape of DataFrame before dropping duplicates based on 'الاسم الثلاثي': {توقيف_df.shape}")

# Drop duplicate rows based on 'الاسم الثلاثي' (Full Name), keeping the first occurrence
توقيف_df = توقيف_df.drop_duplicates(subset=['الاسم الثلاثي'], keep='first')

print(f"Shape of DataFrame after dropping duplicates based on 'الاسم الثلاثي': {توقيف_df.shape}")

# Display the head of the cleaned DataFrame to confirm
display(توقيف_df.head())
```

```python
# Identify and count null values in each column
null_values_count = توقيف_df.isnull().sum()

print("Count of null values per column:")
print(null_values_count)
```

```python
# Check if 'ملاحظات' column exists before trying to drop it
if 'ملاحظات' in توقيف_df.columns:
    توقيف_df = توقيف_df.drop(columns=['ملاحظات'])
    print("DataFrame after dropping 'ملاحظات' column:")
else:
    print("'ملاحظات' column not found in DataFrame. It might have been dropped already.")

# Display the current columns to verify
print("Current columns in DataFrame:")
print(توقيف_df.columns)
display(توقيف_df.head())
```

```python
# Function to extract date using regex
def extract_issue_date(text):
    if pd.isna(text):
        return np.nan
    # Regex to find 'YYYY/MM/DD' or 'YYYY/M/D' pattern after 'تاريخ'
    match = re.search(r'تاريخ\s*(\d{4}/\d{1,2}/\d{1,2})', str(text))
    if match:
        return match.group(1)
    return np.nan

# Apply the function to create the new column
توقيف_df['تاريخ صدور مذكرة التوقيف'] = توقيف_df['المستند القاضي بالتوقيف'].apply(extract_issue_date)

# Convert the new column to datetime format, coercing errors to NaT
توقيف_df['تاريخ صدور مذكرة التوقيف'] = pd.to_datetime(توقيف_df['تاريخ صدور مذكرة التوقيف'], errors='coerce')

print("DataFrame with new 'تاريخ صدور مذكرة التوقيف' column:")
display(توقيف_df.head())
```

```python
# Identify and count null values in each column after creating the new date column
updated_null_values_count = توقيف_df.isnull().sum()

print("Updated count of null values per column:")
print(updated_null_values_count)

# Display data types of all columns to verify date column type
print("\nData types of all columns:")
print(توقيف_df.dtypes)
```

```python
توقيف_df['تاريخ التوقيف'] = pd.to_datetime(توقيف_df['تاريخ التوقيف'], errors='coerce')

print("DataFrame after converting 'تاريخ التوقيف' to datetime:")
display(توقيف_df.head())

print("\nUpdated data types of all columns:")
print(توقيف_df.dtypes)
```
